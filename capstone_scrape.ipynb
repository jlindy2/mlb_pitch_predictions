```python
from pybaseball import statcast
import pandas as pd
```

```python
#scrape 2019 pitch-by-pitch data

df19 = statcast(start_dt='2019-03-25', end_dt='2019-10-30')
```

```python
df19.head()
```

```python
df19.shape
```

```python
df19.info()
```

```python
#scrape 2018 pitch-by-pitch data

df18 = statcast(start_dt='2018-03-25', end_dt='2018-10-30')
```

```python
df18.shape
```

```python
# combine 2018 and 2019

df = pd.concat([df18, df19]).reset_index()
```

```python
df.shape
```

```python
df.tail()
```

```python
df_baseball = df
```

```python
# pickle original dataset
import pickle

df_baseball.to_pickle("df_baseball.pkl")
```

```python
# drop useless columns
df_baseball.drop(['spin_dir', 'spin_rate_deprecated', 'break_angle_deprecated', 'break_length_deprecated', 
         'des', 'tfs_deprecated', 'tfs_zulu_deprecated', 'umpire', 
        'hc_x', 'hc_y', 'level_0', 'index'], axis=1, inplace=True)
```

```python
# drop fielders
df_baseball.drop(['fielder_3', 'fielder_4', 'fielder_5', 'fielder_6', 'fielder_7', 'fielder_8', 
         'fielder_9'], axis=1, inplace=True)
```

```python
# drop hitting results
df_baseball.drop(['hit_distance_sc', 'launch_speed', 'launch_angle', 'effective_speed', 'release_spin_rate', 
        'release_extension', 'release_pos_y', 'estimated_ba_using_speedangle', 
        'estimated_woba_using_speedangle', 'woba_value', 'woba_denom', 'babip_value', 
        'iso_value', 'launch_speed_angle'], axis=1, inplace=True)
```

```python
# drop pitchfx
df_baseball.drop(['release_speed', 'release_pos_x', 'pfx_x', 'pfx_z', 
         'plate_x', 'plate_z', 'vx0', 'vy0', 'vz0', 'ax', 'ay', 'az', 
         'sz_top', 'sz_bot'], axis=1, inplace=True)
```
